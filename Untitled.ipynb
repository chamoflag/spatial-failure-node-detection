```python
import pandas as pd
url="https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
df=pd.read_csv(url,sep='\s+',header=None,names=['mpg','cylinders','displacement','horsepower','weight','acceleration','model_year','origin','car_name'])
print(df.shape)
print(df.head())
print(df.dtypes)
```

```python
print(df['horsepower'].unique())
print(df['horsepower'].value_counts().tail(10))
```

```python
print(df[df['horsepower']=='?'])
```

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
print(df['horsepower'].isna().sum())
print(df.dtypes)
```

```python
missing_hp_idx = df[df['horsepower'].isna()].index.tolist()
print(missing_hp_idx)
```

```python
df['horsepower'].fillna(df['horsepower'].median(), inplace=True)
print(df['horsepower'].isna().sum())
```

```python
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
```

```python
df = df.drop(columns=['car_name'])
print(df.shape)
print(df.head())
```

```python
X = df.drop(columns=['mpg'])
y = df['mpg']

print(X.shape)
print(y.shape)
```

```python
import numpy as np

bins = 10
density_counts = []

for col in X.columns:
    counts, _ = np.histogram(X[col], bins=bins)
    density_counts.append(counts)

density_counts = np.array(density_counts)
cv_per_feature = density_counts.std(axis=1) / density_counts.mean(axis=1)

for col, cv in zip(X.columns, cv_per_feature):
    print(f"{col}: CV = {cv:.3f}")

print(f"\nMean CV across features: {cv_per_feature.mean():.3f}")
```

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

degrees = [1, 2, 3, 4, 5]
models = []

for d in degrees:
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('poly', PolynomialFeatures(degree=d, include_bias=False)),
        ('reg', LinearRegression())
    ])
    pipe.fit(X, y)
    models.append(pipe)
    print(f"Degree {d} fitted")
```

```python
predictions = np.array([model.predict(X) for model in models])

divergence = predictions.std(axis=0)

print(f"Mean divergence: {divergence.mean():.3f}")
print(f"Max divergence: {divergence.max():.3f}")
print(f"Min divergence: {divergence.min():.3f}")
print(f"\nDivergence at ground truth missing indices:")
for idx in missing_hp_idx:
    print(f"  Row {idx}: {divergence[idx]:.3f}")
```

```python
from scipy import stats

# Spearman correlation between divergence and distance from mean prediction
mean_pred = predictions.mean(axis=0)
residuals = np.abs(y - mean_pred)

spearman_corr, spearman_p = stats.spearmanr(divergence, residuals)
print(f"Spearman correlation: {spearman_corr:.3f}")
print(f"Spearman p-value: {spearman_p:.6f}")
```

```python
median_divergence = np.median(divergence)

high_div_residuals = residuals[divergence > median_divergence]
low_div_residuals = residuals[divergence <= median_divergence]

levene_stat, levene_p = stats.levene(high_div_residuals, low_div_residuals)

print(f"Levene statistic: {levene_stat:.3f}")
print(f"Levene p-value: {levene_p:.6f}")
print(f"\nHigh divergence region mean error: {high_div_residuals.mean():.3f}")
print(f"Low divergence region mean error: {low_div_residuals.mean():.3f}")
```

```python
from sklearn.decomposition import PCA
from libpysal.weights import DistanceBand
from esda.moran import Moran
import numpy as np

pca = PCA(n_components=2)
X_2d = pca.fit_transform(X)

print(f"Explained variance ratio: {pca.explained_variance_ratio_}")
print(f"Total variance retained: {pca.explained_variance_ratio_.sum():.3f}")

threshold = np.percentile(
    np.sort(
        np.sqrt(((X_2d[:, None] - X_2d[None, :]) ** 2).sum(axis=2)),
        axis=1
    )[:, 5], 75)

w2d = DistanceBand(X_2d, threshold=threshold,
                   binary=True, silence_warnings=True)

mi2d = Moran(divergence, w2d)
print(f"\nMoran's I (PCA 2D): {mi2d.I:.3f}")
print(f"p-value: {mi2d.p_sim:.6f}")
```

```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/concrete/compressive/Concrete_Data.xls"

df2 = pd.read_excel(url)
print(df2.shape)
print(df2.head())
print(df2.dtypes)
```

```python
X2 = df2.iloc[:, :-1]
y2 = df2.iloc[:, -1]

print(X2.shape)
print(y2.shape)
print(X2.columns.tolist())
```

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from libpysal.weights import KNN
from esda.moran import Moran
from scipy import stats
import numpy as np

# CV check
bins = 10
density_counts = []
for col in X2.columns:
    counts, _ = np.histogram(X2[col], bins=bins)
    density_counts.append(counts)
density_counts = np.array(density_counts)
cv_per_feature = density_counts.std(axis=1) / density_counts.mean(axis=1)
print(f"Mean CV: {cv_per_feature.mean():.3f}")

# ensemble
degrees = [1, 2, 3, 4, 5]
models2 = []
for d in degrees:
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('poly', PolynomialFeatures(degree=d, include_bias=False)),
        ('reg', LinearRegression())
    ])
    pipe.fit(X2, y2)
    models2.append(pipe)

# divergence
predictions2 = np.array([m.predict(X2) for m in models2])
divergence2 = predictions2.std(axis=0)
print(f"Mean divergence: {divergence2.mean():.3f}")
print(f"Max divergence: {divergence2.max():.3f}")

# spearman
mean_pred2 = predictions2.mean(axis=0)
residuals2 = np.abs(y2 - mean_pred2)
spearman_corr2, spearman_p2 = stats.spearmanr(divergence2, residuals2)
print(f"\nSpearman: {spearman_corr2:.3f}, p = {spearman_p2:.6f}")

# levene
median_div2 = np.median(divergence2)
high2 = residuals2[divergence2 > median_div2]
low2 = residuals2[divergence2 <= median_div2]
lev_stat2, lev_p2 = stats.levene(high2, low2)
print(f"Levene: {lev_stat2:.3f}, p = {lev_p2:.6f}")
print(f"High divergence mean error: {high2.mean():.3f}")
print(f"Low divergence mean error: {low2.mean():.3f}")

# morans I
X2_scaled = StandardScaler().fit_transform(X2)
w2 = KNN(X2_scaled, k=5, silence_warnings=True)
mi2 = Moran(divergence2, w2)
print(f"Moran's I: {mi2.I:.3f}, p = {mi2.p_sim:.6f}")
```

```python
import numpy as np
from scipy import stats
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from libpysal.weights import KNN
from esda.moran import Moran

def run_pipeline(X_in, y_in, label):
    degrees = [1, 2, 3, 4, 5]
    models_temp = []
    for d in degrees:
        pipe = Pipeline([
            ('scaler', StandardScaler()),
            ('poly', PolynomialFeatures(degree=d, include_bias=False)),
            ('reg', LinearRegression())
        ])
        pipe.fit(X_in, y_in)
        models_temp.append(pipe)

    preds = np.array([m.predict(X_in) for m in models_temp])
    div = preds.std(axis=0)
    mean_pred = preds.mean(axis=0)
    resid = np.abs(y_in - mean_pred)

    sp_corr, sp_p = stats.spearmanr(div, resid)

    median_div = np.median(div)
    high = resid[div > median_div]
    low = resid[div <= median_div]
    lev_stat, lev_p = stats.levene(high, low)

    X_scaled = StandardScaler().fit_transform(X_in)
    w = KNN(X_scaled, k=5, silence_warnings=True)
    mi = Moran(div, w)

    print(f"\n--- {label} ---")
    print(f"Spearman: {sp_corr:.3f}, p = {sp_p:.6f}")
    print(f"Levene: {lev_stat:.3f}, p = {lev_p:.6f}")
    print(f"High/Low error: {high.mean():.3f} / {low.mean():.3f}")
    print(f"Moran's I: {mi.I:.3f}, p = {mi.p_sim:.6f}")

# Real Auto MPG
run_pipeline(X.values, y.values, "REAL DATA - Auto MPG")

# Shuffle test - same X, shuffled y
y_shuffled = y.values.copy()
np.random.shuffle(y_shuffled)
run_pipeline(X.values, y_shuffled, "SHUFFLE TEST - y randomized")

# Pure random baseline
np.random.seed(42)
X_random = np.random.randn(398, 7)
y_random = np.random.randn(398)
run_pipeline(X_random, y_random, "RANDOM BASELINE - no real structure")
```
