```python
import pandas as pd
url="https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
df=pd.read_csv(url,sep='\s+',header=None,names=['mpg','cylinders','displacement','horsepower','weight','acceleration','model_year','origin','car_name'])
print(df.shape)
print(df.head())
print(df.dtypes)
```

```python
df = df.drop(columns=['car_name'])
print(df.shape)
print(df.head())
```

```python
X = df.drop(columns=['mpg'])
y = df['mpg']

print(X.shape)
print(y.shape)
```

```python
import numpy as np

bins = 10
density_counts = []

for col in X.columns:
    counts, _ = np.histogram(X[col], bins=bins)
    density_counts.append(counts)

density_counts = np.array(density_counts)
cv_per_feature = density_counts.std(axis=1) / density_counts.mean(axis=1)

for col, cv in zip(X.columns, cv_per_feature):
    print(f"{col}: CV = {cv:.3f}")

print(f"\nMean CV across features: {cv_per_feature.mean():.3f}")
```

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

degrees = [1, 2, 3, 4, 5]
models = []

for d in degrees:
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('poly', PolynomialFeatures(degree=d, include_bias=False)),
        ('reg', LinearRegression())
    ])
    pipe.fit(X, y)
    models.append(pipe)
    print(f"Degree {d} fitted")
```

```python
predictions = np.array([model.predict(X) for model in models])

divergence = predictions.std(axis=0)

print(f"Mean divergence: {divergence.mean():.3f}")
print(f"Max divergence: {divergence.max():.3f}")
print(f"Min divergence: {divergence.min():.3f}")
print(f"\nDivergence at ground truth missing indices:")
for idx in missing_hp_idx:
    print(f"  Row {idx}: {divergence[idx]:.3f}")
```

```python
from scipy import stats

# Spearman correlation between divergence and distance from mean prediction
mean_pred = predictions.mean(axis=0)
residuals = np.abs(y - mean_pred)

spearman_corr, spearman_p = stats.spearmanr(divergence, residuals)
print(f"Spearman correlation: {spearman_corr:.3f}")
print(f"Spearman p-value: {spearman_p:.6f}")
```

```python
median_divergence = np.median(divergence)

high_div_residuals = residuals[divergence > median_divergence]
low_div_residuals = residuals[divergence <= median_divergence]

levene_stat, levene_p = stats.levene(high_div_residuals, low_div_residuals)

print(f"Levene statistic: {levene_stat:.3f}")
print(f"Levene p-value: {levene_p:.6f}")
print(f"\nHigh divergence region mean error: {high_div_residuals.mean():.3f}")
print(f"Low divergence region mean error: {low_div_residuals.mean():.3f}")
```

```python
import numpy as np
from scipy import stats
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from libpysal.weights import KNN
from esda.moran import Moran

def run_pipeline(X_in, y_in, label):
    degrees = [1, 2, 3, 4, 5]
    models_temp = []
    for d in degrees:
        pipe = Pipeline([
            ('scaler', StandardScaler()),
            ('poly', PolynomialFeatures(degree=d, include_bias=False)),
            ('reg', LinearRegression())
        ])
        pipe.fit(X_in, y_in)
        models_temp.append(pipe)

    preds = np.array([m.predict(X_in) for m in models_temp])
    div = preds.std(axis=0)
    mean_pred = preds.mean(axis=0)
    resid = np.abs(y_in - mean_pred)

    sp_corr, sp_p = stats.spearmanr(div, resid)

    median_div = np.median(div)
    high = resid[div > median_div]
    low = resid[div <= median_div]
    lev_stat, lev_p = stats.levene(high, low)

    X_scaled = StandardScaler().fit_transform(X_in)
    w = KNN(X_scaled, k=5, silence_warnings=True)
    mi = Moran(div, w)

    print(f"\n--- {label} ---")
    print(f"Spearman: {sp_corr:.3f}, p = {sp_p:.6f}")
    print(f"Levene: {lev_stat:.3f}, p = {lev_p:.6f}")
    print(f"High/Low error: {high.mean():.3f} / {low.mean():.3f}")
    print(f"Moran's I: {mi.I:.3f}, p = {mi.p_sim:.6f}")

# Real Auto MPG
run_pipeline(X.values, y.values, "REAL DATA - Auto MPG")

# Shuffle test - same X, shuffled y
y_shuffled = y.values.copy()
np.random.shuffle(y_shuffled)
run_pipeline(X.values, y_shuffled, "SHUFFLE TEST - y randomized")

# Pure random baseline
np.random.seed(42)
X_random = np.random.randn(398, 7)
y_random = np.random.randn(398)
run_pipeline(X_random, y_random, "RANDOM BASELINE - no real structure")
```

```python
import numpy as np
import pandas as pd
import logging
from typing import Tuple, Dict, Any, List
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from scipy import stats
import libpysal
from esda.moran import Moran
from libpysal.weights import DistanceBand

# Configure professional logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger("FailureNodeDetector")

class EnsembleTopographyDetector:
    """
    A framework to algorithmically map, validate, and stress-test
    localized failure nodes in machine learning feature spaces.
    """
    def __init__(self, degrees: List[int] = [1, 2, 3, 4, 5], knn_neighbors: int = 6):
        self.degrees = degrees
        self.knn_neighbors = knn_neighbors
        self.models = []
        self.is_fitted = False

    def fit(self, X: pd.DataFrame, y: pd.Series) -> 'EnsembleTopographyDetector':
        """Trains the structural ensemble across varying polynomial complexities."""
        logger.info(f"Fitting ensemble of polynomial models for degrees: {self.degrees}")
        self.models = []

        for d in self.degrees:
            pipe = Pipeline([
                ('scaler', StandardScaler()),
                ('poly', PolynomialFeatures(degree=d, include_bias=False)),
                ('reg', LinearRegression())
            ])
            pipe.fit(X, y)
            self.models.append(pipe)

        self.is_fitted = True
        return self

    def compute_metrics(self, X: pd.DataFrame, y: pd.Series) -> Dict[str, Any]:
        """Calculates divergence topography, error alignment, and spatial clustering."""
        if not self.is_fitted:
            raise ValueError("The detector instance must be fitted before computing metrics.")

        X_arr = X.values
        y_arr = y.values

        # 1. Compute Ensemble Predictions & Divergence (Epistemic Uncertainty)
        predictions = np.array([model.predict(X) for model in self.models])
        mean_predictions = predictions.mean(axis=0)
        divergence = predictions.std(axis=0)
        residuals = np.abs(y_arr - mean_predictions)

        # 2. Statistical Alignment (Spearman Rank Correlation)
        spearman_corr, spearman_p = stats.spearmanr(divergence, residuals)

        # 3. Variance Disparity Test (Levene's Test)
        median_div = np.median(divergence)
        high_div_res = residuals[divergence > median_div]
        low_div_res = residuals[divergence <= median_div]
        levene_stat, levene_p = stats.levene(high_div_res, low_div_res)

        # 4. Spatial Autocorrelation (Moran's I over Scaled Feature Space)
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X_arr)

        # Determine distance threshold using KDTree structure
        kd = libpysal.cg.KDTree(X_scaled)
        distances, _ = kd.query(X_scaled, k=self.knn_neighbors)
        threshold = np.percentile(distances[:, -1], 75)

        # Calculate spatial weight matrix
        w = DistanceBand(X_scaled, threshold=threshold, binary=True, silence_warnings=True)
        w.transform = 'R' # Row-standardized weights

        moran = Moran(divergence, w)

        return {
            "divergence": divergence,
            "residuals": residuals,
            "spearman_corr": spearman_corr,
            "spearman_p": spearman_p,
            "levene_stat": levene_stat,
            "levene_p": levene_p,
            "moran_i": moran.I,
            "moran_p": moran.p_sim,
            "high_div_mean_error": high_div_res.mean(),
            "low_div_mean_error": low_div_res.mean()
        }

    def run_heteroscedastic_stress_test(self, X: pd.DataFrame, y: pd.Series,
                                        mask_condition: np.ndarray,
                                        noise_std: float = 25.0) -> Dict[str, Any]:
        """
        Injects severe aleatoric noise into a localized region of the target variable.
        Tests if the framework can separate pure noise from structural model failure.
        """
        logger.info("Executing Stress Test: Injecting localized heteroscedastic noise...")

        # Induce aleatoric pocket noise
        y_stressed = y.copy().astype(float)
        noise = np.random.normal(0, noise_std, size=mask_condition.sum())
        y_stressed[mask_condition] += noise

        # Re-evaluate the pipeline on corrupted data
        stress_detector = EnsembleTopographyDetector(degrees=self.degrees, knn_neighbors=self.knn_neighbors)
        stress_detector.fit(X, y_stressed)
        stress_metrics = stress_detector.compute_metrics(X, y_stressed)

        return stress_metrics

# ==========================================
# EXECUTION PIPELINE
# ==========================================
if __name__ == "__main__":
    # Load and clean Auto MPG dataset
    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
    df = pd.read_csv(url, sep='\s+', header=None,
                     names=['mpg', 'cylinders', 'displacement', 'horsepower',
                            'weight', 'acceleration', 'model_year', 'origin', 'car_name'])

    df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
    df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
    X = df.drop(columns=['mpg', 'car_name'])
    y = df['mpg']

    # Initialize and evaluate the production-grade pipeline
    detector = EnsembleTopographyDetector(degrees=[1, 2, 3, 4, 5])
    baseline_metrics = detector.fit(X, y).compute_metrics(X, y)

    print("\n" + "="*50)
    print("BASELINE REAL DATA RESULTS")
    print("="*50)
    print(f"Spearman Correlation : {baseline_metrics['spearman_corr']:.3f} (p: {baseline_metrics['spearman_p']:.6f})")
    print(f"Levene Statistic     : {baseline_metrics['levene_stat']:.3f} (p: {baseline_metrics['levene_p']:.6f})")
    print(f"Moran's I (Spatial)  : {baseline_metrics['moran_i']:.3f} (p: {baseline_metrics['moran_p']:.6f})")
    print(f"High-Div Region Error: {baseline_metrics['high_div_mean_error']:.3f}")
    print(f"Low-Div Region Error : {baseline_metrics['low_div_mean_error']:.3f}")

    # STRESS TEST: Inject huge random noise ONLY for heavy cars (weight > 4000 lbs)
    # If Moran's I drops significantly, it proves the detector isolates structural failure,
    # not just random environment noise.
    heavy_cars_mask = (X['weight'] > 4000).values
    stress_metrics = detector.run_heteroscedastic_stress_test(X, y, mask_condition=heavy_cars_mask, noise_std=30.0)

    print("\n" + "="*50)
    print("STRESS TEST: LOCALIZED HETEROSCEDASTIC NOISE RESULTS")
    print("="*50)
    print(f"Spearman Correlation : {stress_metrics['spearman_corr']:.3f} (p: {stress_metrics['spearman_p']:.6f})")
    print(f"Levene Statistic     : {stress_metrics['levene_stat']:.3f} (p: {stress_metrics['levene_p']:.6f})")
    print(f"Moran's I (Spatial)  : {stress_metrics['moran_i']:.3f} (p: {stress_metrics['moran_p']:.6f})")
    print(f"High-Div Region Error: {stress_metrics['high_div_mean_error']:.3f}")
    print(f"Low-Div Region Error : {stress_metrics['low_div_mean_error']:.3f}")
```

```python
import numpy as np
import pandas as pd
import logging
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Tuple, Dict, Any, List
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from scipy import stats
import libpysal
from esda.moran import Moran
from libpysal.weights import DistanceBand

# Configure professional logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger("FailureNodeDetector")

class EnsembleTopographyDetector:
    """
    A framework to algorithmically map, validate, and visualize
    localized failure nodes in machine learning feature spaces.
    """
    def __init__(self, degrees: List[int] = [1, 2, 3, 4, 5], knn_neighbors: int = 6):
        self.degrees = degrees
        self.knn_neighbors = knn_neighbors
        self.models = []
        self.is_fitted = False

    def fit(self, X: pd.DataFrame, y: pd.Series) -> 'EnsembleTopographyDetector':
        """Trains the structural ensemble across varying polynomial complexities."""
        logger.info(f"Fitting ensemble of polynomial models for degrees: {self.degrees}")
        self.models = []

        for d in self.degrees:
            pipe = Pipeline([
                ('scaler', StandardScaler()),
                ('poly', PolynomialFeatures(degree=d, include_bias=False)),
                ('reg', LinearRegression())
            ])
            pipe.fit(X, y)
            self.models.append(pipe)

        self.is_fitted = True
        return self

    def compute_metrics(self, X: pd.DataFrame, y: pd.Series) -> Dict[str, Any]:
        """Calculates divergence topography, error alignment, and spatial clustering."""
        if not self.is_fitted:
            raise ValueError("The detector instance must be fitted before computing metrics.")

        X_arr = X.values
        y_arr = y.values

        # 1. Compute Ensemble Predictions & Divergence (Epistemic Uncertainty)
        predictions = np.array([model.predict(X) for model in self.models])
        mean_predictions = predictions.mean(axis=0)
        divergence = predictions.std(axis=0)
        residuals = np.abs(y_arr - mean_predictions)

        # 2. Statistical Alignment (Spearman Rank Correlation)
        spearman_corr, spearman_p = stats.spearmanr(divergence, residuals)

        # 3. Variance Disparity Test (Levene's Test)
        median_div = np.median(divergence)
        high_div_res = residuals[divergence > median_div]
        low_div_res = residuals[divergence <= median_div]
        levene_stat, levene_p = stats.levene(high_div_res, low_div_res)

        # 4. Spatial Autocorrelation (Moran's I over Scaled Feature Space)
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X_arr)

        kd = libpysal.cg.KDTree(X_scaled)
        distances, _ = kd.query(X_scaled, k=self.knn_neighbors)
        threshold = np.percentile(distances[:, -1], 75)

        w = DistanceBand(X_scaled, threshold=threshold, binary=True, silence_warnings=True)
        w.transform = 'R'

        moran = Moran(divergence, w)

        return {
            "divergence": divergence,
            "residuals": residuals,
            "spearman_corr": spearman_corr,
            "spearman_p": spearman_p,
            "levene_stat": levene_stat,
            "levene_p": levene_p,
            "moran_i": moran.I,
            "moran_p": moran.p_sim,
            "high_div_mean_error": high_div_res.mean(),
            "low_div_mean_error": low_div_res.mean()
        }

    def run_heteroscedastic_stress_test(self, X: pd.DataFrame, y: pd.Series,
                                        mask_condition: np.ndarray,
                                        noise_std: float = 25.0) -> Tuple[Dict[str, Any], pd.Series]:
        """
        Injects severe aleatoric noise into a localized region of the target variable.
        """
        logger.info("Executing Stress Test: Injecting localized heteroscedastic noise...")
        y_stressed = y.copy().astype(float)
        noise = np.random.normal(0, noise_std, size=mask_condition.sum())
        y_stressed[mask_condition] += noise

        stress_detector = EnsembleTopographyDetector(degrees=self.degrees, knn_neighbors=self.knn_neighbors)
        stress_detector.fit(X, y_stressed)
        stress_metrics = stress_detector.compute_metrics(X, y_stressed)

        return stress_metrics, y_stressed

    def plot_topography(self, X: pd.DataFrame, metrics: Dict[str, Any],
                        feature_x: str, feature_y: str, title_suffix: str = "Baseline") -> None:
        """
        Generates an advanced spatial map displaying model divergence
        and explicitly isolates algorithmic failure nodes.
        """
        divergence = metrics["divergence"]

        # Define severe failure nodes as points in the top 10% of model disagreement
        failure_threshold = np.percentile(divergence, 90)
        failure_nodes_mask = divergence >= failure_threshold

        sns.set_theme(style="whitegrid")
        fig, axes = plt.subplots(1, 2, figsize=(18, 7), sharex=True, sharey=True)

        # Plot 1: Continuous Uncertainty Landscape
        sc1 = axes[0].scatter(X[feature_x], X[feature_y], c=divergence, cmap='plasma',
                             alpha=0.8, edgecolors='none', s=50)
        cbar1 = fig.colorbar(sc1, ax=axes[0])
        cbar1.set_label('Model Divergence ($\sigma_{ensemble}$)', fontsize=12)
        axes[0].set_title(f'Uncertainty Landscape ({title_suffix})', fontsize=14, fontweight='bold')
        axes[0].set_xlabel(feature_x.capitalize(), fontsize=12)
        axes[0].set_ylabel(feature_y.capitalize(), fontsize=12)

        # Plot 2: Isolated Algorithmic Failure Nodes
        axes[1].scatter(X[feature_x], X[feature_y], color='lightgray', alpha=0.5, s=30, label='Stable Regions')
        axes[1].scatter(X.loc[failure_nodes_mask, feature_x], X.loc[failure_nodes_mask, feature_y],
                        color='crimson', alpha=0.9, edgecolors='black', s=70, linewidths=1.2,
                        label='Isolated Failure Nodes (Top 10%)')

        axes[1].set_title(f'Algorithmic Failure Nodes ({title_suffix})', fontsize=14, fontweight='bold')
        axes[1].set_xlabel(feature_x.capitalize(), fontsize=12)
        axes[1].legend(loc='upper right', frameon=True, facecolor='white', edgecolor='none')

        suptitle = fig.suptitle(f"Spatial Error Topography Analysis — Moran's I: {metrics['moran_i']:.3f}",
                                    fontsize=16, fontweight='bold', y=1.02)

        plt.tight_layout()
        plt.show()


# ==========================================
# EXECUTION PIPELINE
# ==========================================
if __name__ == "__main__":
    # Load and clean Auto MPG dataset (Fixed raw string regex warning)
    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
    df = pd.read_csv(url, sep=r'\s+', header=None,
                     names=['mpg', 'cylinders', 'displacement', 'horsepower',
                            'weight', 'acceleration', 'model_year', 'origin', 'car_name'])

    df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
    df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
    X = df.drop(columns=['mpg', 'car_name'])
    y = df['mpg']

    # 1. Run Baseline Real Data Pipeline
    detector = EnsembleTopographyDetector(degrees=[1, 2, 3, 4, 5])
    baseline_metrics = detector.fit(X, y).compute_metrics(X, y)

    print("\n" + "="*50)
    print("BASELINE REAL DATA RESULTS")
    print("="*50)
    print(f"Moran's I (Spatial)  : {baseline_metrics['moran_i']:.3f} (p: {baseline_metrics['moran_p']:.6f})")
    print(f"High-Div Region Error: {baseline_metrics['high_div_mean_error']:.3f}")

    # Plot Baseline Map (Looking at Weight vs Horsepower layout)
    detector.plot_topography(X, baseline_metrics, feature_x='weight', feature_y='horsepower', title_suffix="Baseline")

    # 2. Run Localized Stress Test (Inject noise for heavy cars > 4000 lbs)
    heavy_cars_mask = (X['weight'] > 4000).values
    stress_metrics, y_stressed = detector.run_heteroscedastic_stress_test(X, y, mask_condition=heavy_cars_mask, noise_std=30.0)

    print("\n" + "="*50)
    print("STRESS TEST: LOCALIZED HETEROSCEDASTIC NOISE RESULTS")
    print("="*50)
    print(f"Moran's I (Spatial)  : {stress_metrics['moran_i']:.3f} (p: {stress_metrics['moran_p']:.6f})")
    print(f"High-Div Region Error: {stress_metrics['high_div_mean_error']:.3f}")

    # Plot Stress Test Map (Watch the failure nodes instantly cluster in the upper-right high weight region)
    detector.plot_topography(X, stress_metrics, feature_x='weight', feature_y='horsepower', title_suffix="Noise Stress Test")
```
